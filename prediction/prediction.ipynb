```python
import pandas as pd
import numpy as np
```

```python
test= pd.read_csv("../data/Test.csv")
test.head()
```

```python
test.isnull().sum()
```

```python
test['Item_Weight']=test['Item_Weight'].fillna(test.groupby('Item_Identifier')['Item_Weight'].transform('mean'))
test.isnull().sum()
```

```python
test['Item_Weight']=test['Item_Weight'].fillna(test.groupby('Item_Type')['Item_Weight'].transform('mean'))
test.isnull().sum()
```

```python
pd.crosstab(test.Outlet_Size,test.Outlet_Location_Type)
```

```python
dic= {"Tier 2":"Small"}
s= test["Outlet_Location_Type"].map(dic)

test.Outlet_Size = test.Outlet_Size.combine_first(s)

test.Outlet_Size.value_counts()
```

```python
test.isnull().sum()
```

```python
pd.crosstab(test.Outlet_Size,test.Outlet_Type)
```

```python
dic = {'Grocery Store':'Small'}
s = test.Outlet_Type.map(dic)
test.Outlet_Size= test.Outlet_Size.combine_first(s)
test.isnull().sum()
```

```python
## Outlet_Years

test["Outlet_Years"]= test["Outlet_Establishment_Year"].apply(lambda x: 2021-x)
test.head()
```

```python
# Item_Type_New

def get_first_two_char(obj):
    first_two= obj[:2]
    if first_two== "FD":
        return "Food"
    elif first_two =="DR":
        return "Drinks"
    else:
        return "Non-Consumable"

test["Item_Type_New"]= test["Item_Identifier"].apply(get_first_two_char)
```

```python
test.head()
```

```python
test.Item_Fat_Content.unique()
```

```python
test["Item_Fat_Content"]= test["Item_Fat_Content"].map(
    {"low fat":"Low Fat",
     "reg":"Regular Fat",
     "LF":"Low Fat",
     "Regular":"Regular Fat",
     "Low Fat":"Low Fat"})
test.Item_Fat_Content.unique()
```

```python
imputed_test= test.copy()
```

### features that to be label encoded
1. Item_Fat_Content 
2. Outlet_Identifier
3. Outlet_Size
4. Outlet_Location_Type
5. Outlet_Type
6. Item_Type
7. Item_Type_New

```python
import pickle

labels= ["Item_Fat_Content","Outlet_Identifier","Outlet_Size","Outlet_Location_Type","Outlet_Type","Item_Type","Item_Type_New"]

for label in labels:
    try:
        filename= "le_" + label + ".sav"
        encoder = pickle.load(open("../cleaning/Label_encoder/"+filename,"rb"))
        test[label]= encoder.transform(test[label])
    except Exception as e:
        raise e
        
```

```python
test.head()
```

```python
test= test.drop(columns=["Item_Identifier","Outlet_Establishment_Year"],axis= 1)
test.head()
```

## Prediction

### scaling

```python
with open("../model_training/scaler.sav","rb") as f:
    scaler = pickle.load(f)
```

```python
test= scaler.transform(test)
```

### loading model

```python
with open("../model_training/model.sav","rb") as f:
    model = pickle.load(f)
```

```python
predicted_sales= model.predict(test)
```

```python
imputed_test["Predicted_Sales"]= predicted_sales
imputed_test.head(10)
```
